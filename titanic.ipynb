```python
from sklearn.impute import SimpleImputer

# Seleccionamos solo la columna 'Age'
age = df_train[['Age']]

# Creamos un objeto imputador
imputer = SimpleImputer(strategy='median')

# Ajustamos el imputador al conjunto de entrenamiento
imputer.fit(age)

# Transformamos el conjunto de entrenamiento
age_imputed = imputer.transform(age)

# Volvemos a crear un dataframe de Pandas
df_train['Age'] = age_imputed
```

```python
# Obtenemos el valor más común de 'Embarked'
most_common = df_train['Embarked'].value_counts().index[0]

# Reemplazamos los valores faltantes con el valor más común
df_train['Embarked'].fillna(most_common, inplace=True)
```

```python
# Creamos las variables dummy para 'Sex'
sex_dummies = pd.get_dummies(df_train['Sex'], prefix='Sex')

# Añadimos las variables dummy al dataframe
df_train = pd.concat([df_train, sex_dummies], axis=1)

# Eliminamos la columna 'Sex'
df_train.drop('Sex', axis=1, inplace=True)
```
